### 数据透视表

```python
# 读取本地表格数据
data=pd.read_excel(path)
data.head(5)
```

```python
# pivot_table
# 运算今天的所有数据
data.pivot_table('open',index='date',columns='close',aggfunc='mean').fillna(0)
```

```python
# 分类求和
data.pivot_table('','')
```

```python
# 分类计算，最大值/最小值
def rateFunc(group):
    return group.max()/group.min()

data.pivot_table('open',index='date',columns='close',aggfunc=rateFunc).fillna(0)
```
